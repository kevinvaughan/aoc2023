```python
class Graph:
    def __init__(self):
        self.nodes = {}
        self.memo = {}  # Key: (current_node, remaining_instructions), Value: (end_node, steps)

    def add_node(self, name, left, right):
        self.nodes[name] = (left, right)

    def traverse(self, start, instructions):
        current = start
        total_steps = 0

        while current != 'ZZZ':
            if (current, instructions) in self.memo:
                end_node, steps = self.memo[(current, instructions)]
                total_steps += steps
                current = end_node
                continue

            steps, next_node = self.apply_instructions(current, instructions)
            self.memo[(current, instructions)] = (next_node, steps)
            total_steps += steps
            current = next_node

        return total_steps

    def apply_instructions(self, start, instructions):
        current = start
        steps = 0

        for instruction in instructions:
            current = self.nodes[current][0 if instruction == 'L' else 1]
            steps += 1

            if current == 'ZZZ':
                break

        return steps, current
```

```python
src = input.copy()

# Example Usage
graph = Graph()

instructions = src[0]
for node in src[2:]:
    # LRV = (NNC, BHD)
    name, branches = node.split(' = ')
    left, right = branches[1:-1].split(', ')
    graph.add_node(name, left, right)

print(f"Instructions: {instructions}")

steps = graph.traverse('AAA', instructions)
print(f"Steps taken: {steps}")

print(graph.memo)
```
